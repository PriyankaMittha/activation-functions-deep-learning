# 🔥 Activation Functions and Their Derivatives

This notebook demonstrates different activation functions used in deep learning along with their derivatives and visualizations.

## 📌 Functions Covered:
- Sigmoid
- ReLU
- Tanh
- Leaky ReLU
- Parametric ReLU
- Softmax

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')
```

## 🔹 Sigmoid Function

- Range: (0, 1)
- Used in: Binary Classification
- Problem: Vanishing Gradient

```python
# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of sigmoid activation function
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

## 🔹 ReLU Function

- Range: [0, ∞)
- Most used activation function
- Advantage: Fast and simple
- Problem: Dead neurons

```python
# ReLU activation function
def relu(x):
    return np.maximum(0,x)

# Derivative of ReLU activation function
def relu_derivative(x):
    return np.where(x <= 0, 0, 1)
```

## 🔹 Tanh Function

- Range: (-1, 1)
- Zero-centered
- Still suffers from vanishing gradient

```python
# Hyperbolic Tangent (tanh) activation function
def tanh(x):
    return np.tanh(x)

# Derivative of tanh activation function
def tanh_derivative(x):
    return 1 - np.square(tanh(x))
```

## 🔹 Leaky ReLU

- Fixes dying ReLU problem
- Allows small gradient for negative values

```python
# Leaky ReLU activation function
def leaky_relu(x, alpha = 0.01):
    return np.where(x > 0, x, alpha * x)

# Derivative of Leaky ReLU activation function
def leaky_relu_derivative(x, alpha = 0.01):
    dx = np.ones_like(x)
    dx[x < 0] = alpha
    return dx
```

## 🔹 Parametric ReLU (PReLU) alpha = 0.01

- Alpha is learnable parameter (not fixed)
- Used in advanced models

```python
# Parametric ReLU activation function
def parametric_relu(x, alpha = 0.01):
    return np.where(x > 0, x, alpha * x)

# Derivative of Parametric ReLU activation function
def parametric_relu_derivative(x, alpha = 0.01):
    return np.where(x > 0, 1, alpha)
```

# Softmax 

- Softmax is an activation function used for multi-class classification
- It converts numbers into probabilities (values between 0 and 1) and their sum = 1

```python
# Softmax activation function
def softmax(x):
    exp_vals = np.exp(x - np.max(x))
    return exp_vals / np.sum(exp_vals)
```

```python
# Input values
X = np.linspace(-5, 5, 100)
```

```python
# Compute activation functions
sigmoid_y = sigmoid(X)
relu_y = relu(X)
tanh_y = tanh(X)
leaky_relu_y = leaky_relu(X)
parametric_relu_y = parametric_relu(X)
softmax_y = softmax(np.array([X, X/2, X*2]))
```

```python
# Compute derivatives
sigmoid_derivative_y = sigmoid_derivative(X)
relu_derivative_y = relu_derivative(X)
tanh_derivative_y = tanh_derivative(X)
leaky_relu_derivative_y = leaky_relu_derivative(X)
parametric_relu_derivative_y = parametric_relu_derivative(X)
```

```python
# Plotting the activation functions and their derivatives
plt.figure(figsize=(12,8))

# Sigmoid
plt.subplot(2,3,1)
plt.plot(X, sigmoid_y, label = 'Sigmoid')
plt.plot(X, sigmoid_derivative_y, label = 'Sigmoid Derivative')
plt.title('Sigmoid and Derivative')
plt.legend()

# ReLU
plt.subplot(2,3,2)
plt.plot(X, relu_y, label = 'ReLU')
plt.plot(X, relu_derivative_y, label = 'ReLU Derivative')
plt.title('ReLU and Derivative')
plt.legend()

# Tanh
plt.subplot(2,3,3)
plt.plot(X, tanh_y, label = 'Tanh')
plt.plot(X, tanh_derivative_y, label = 'Tanh Derivative')
plt.title('Tanh and Derivative')
plt.legend()

# Leaky ReLU
plt.subplot(2,3,4)
plt.plot(X, leaky_relu_y, label = 'Leaky ReLU')
plt.plot(X, leaky_relu_derivative_y, label = 'Leaky ReLU Derivative')
plt.title('Leaky ReLU and Derivative')
plt.legend()

# Parametric ReLU
plt.subplot(2,3,5)
plt.plot(X, parametric_relu_y, label = 'Parametric ReLU')
plt.plot(X, parametric_relu_derivative_y, label = 'Parametric ReLU Derivative')
plt.title('Parametric ReLU and Derivative')
plt.legend()

# softmax
plt.subplot(2,3,6)
plt.plot(X, softmax_y.T)
plt.title('Softmax')

plt.tight_layout()
plt.show()
```

## 🎯 Key Takeaways

- ReLU is most widely used activation function
- Sigmoid suffers from vanishing gradient
- Tanh is zero-centered but still has gradient issues
- Leaky ReLU solves dead neuron problem
- Softmax is used for multi-class classification
